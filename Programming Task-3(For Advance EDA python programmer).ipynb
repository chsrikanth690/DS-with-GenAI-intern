#### Maximum Product of Two Elements in an Array

```python
class Solution:
    def maxProduct(self, nums: list[int]) -> int:
        nums.sort()
        return (nums[-1] - 1) * (nums[-2] - 1)

# test
sol = Solution()
print(sol.maxProduct([3, 4, 5, 2])) 
```

#### Count Number of Teams

```python
from typing import List
```

```python
class Solution:
    def numTeams(self, rating: List[int]) -> int:
        n = len(rating)
        teams = 0

        for j in range(n):
            left_smaller = left_greater = 0
            right_smaller = right_greater = 0

            # Count left side
            for i in range(j):
                if rating[i] < rating[j]:
                    left_smaller += 1
                else:
                    left_greater += 1

            # Count right side
            for k in range(j + 1, n):
                if rating[k] > rating[j]:
                    right_greater += 1
                else:
                    right_smaller += 1

            teams += left_smaller * right_greater
            teams += left_greater * right_smaller

        return teams
# Run the code
sol = Solution()
print(sol.numTeams([2, 5, 3, 4, 1]))
        
```

#### Number of Students Doing Homework at a Given Time

```python
class Solution:
    def busyStudent(self, startTime, endTime, queryTime):
        count = 0
        for i in range(len(startTime)):
            if queryTime in list(range(startTime[i],endTime[i]+1)):
                count+=1
        return count
obj = Solution()
obj.busyStudent([1,2,4],[6,7,8],4)
```

#### Number of Steps to Reduce a Number to Zero

```python
class Solution:
    def numberOfSteps(self, num: int) -> int:
        count = 0
        while num != 0:
            if num % 2 == 0:
                num //= 2
            else:
                num -= 1
            count += 1
        return count

# Run the code
sol = Solution()
print(sol.numberOfSteps(14))
```

#### Counting Bits

```python
class Solution:
    def countBits(self, n: int) -> List[int]:
        ans = [0] * (n + 1)
        for i in range(1, n + 1):
            ans[i] = ans[i >> 1] + (i & 1)

        return ans

# Run the code
sol = Solution()
print(sol.countBits(5))
```
